# Find and Save the Bulletins

```python
import pandas as pd

df = pd.read_csv("/Users/shreyasguha/Downloads/Shreyas/code/FinanceGroup2/Monetary Policy Pre-training Dataset - RBIBulletin.csv")
df.head()
```

```python
urls = []

for i in range(len(df)):
    urls.append(df.loc[i]['URL'])
```

```python
import requests
import os

# Make sure the directory exists
save_dir = "/Users/shreyasguha/Downloads/Shreyas/code/FinanceGroup2/Bulletins"
# os.makedirs(save_dir, exist_ok=True)


for idx, url in enumerate(urls):
    if(idx >= 2):
        try:
            response = requests.get(url)
            response.raise_for_status()  # Raise error if request failed

            # Define the full file path
            filename = os.path.join(save_dir, f"file_{idx+1}.pdf")

            with open(filename, 'wb') as f:
                f.write(response.content)
            
            print(f"Downloaded: {filename}")

        except requests.exceptions.RequestException as e:
            print(f"Failed to download {url}: {e}")
```

# Extract text from the bulletins

```python
from unstructured.partition.pdf import partition_pdf

def extract(filepath, i, data):
    elements = partition_pdf(filename=filepath,
                            strategy="fast",
                            extract_images_in_pdf=False,
                            )

    # text_elements = [element for element in elements if element.category == "NarrativeText"]
    print("Number of elements: ", len(elements))

    text_elements = []
    for element in elements:
        if element.category == "NarrativeText" or element.category == "ListItem":
            text_elements.append(element)

    print("Number of text elements: ", len(text_elements))
    text = ""

    for element in text_elements:
        text += element.text
        text += " "

    print("Length of the text: ", len(text))
    # i=0
    # while(i<len(text)):
    #     print(text[i: i+150])
    #     i += 150

    index_name = "file" + str(i)
    data[index_name] = text
    len(data)
    return data
```

```python
import os

folder_path = "/Users/shreyasguha/Downloads/Shreyas/code/FinanceGroup2/PDFs/MPC_report_international/USA"

for i, filename in enumerate(os.listdir(folder_path)):
    if os.path.isfile(os.path.join(folder_path, filename)):
        full_filepath = folder_path+ "/" + filename
        # print(full_filepath, i)
        print(i)
        data = extract(full_filepath, i, data) 
        print("\n\n")
```
